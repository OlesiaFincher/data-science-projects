## Деление данных на выбоки

Разделим данные на обучающую, валидационную и тестовую выборки в пропорциях (60%, 20% и 20%)

```python
# деление данных на обучающую, валидационную и тестовую выборки
train, df_rest = train_test_split(df, test_size=0.4, random_state=12345)
valid, test = train_test_split(df_rest, test_size=0.5, random_state=12345)

print(train.shape)
print(valid.shape)
print(test.shape)
```

Создадим переменные для признаков и целевого признака для каждой выборки

```python
# создание переменных для признаков и целевого признака для обучающей выборки
train_features = train.drop('is_ultra', axis=1)
train_target = train['is_ultra']

# создание переменных для признаков и целевого признака для валидационной выборки
valid_features = valid.drop('is_ultra', axis=1)
valid_target = valid['is_ultra']

# создание переменных для признаков и целевого признака для тестовой выборки
test_features = test.drop('is_ultra', axis=1)
test_target = test['is_ultra']
```

## Исследование моделей

### Решающее дерево

Сохраним в переменную <b>model</b> алгоритм обучения решающее дерево. Затем обучим модель. Для оценки качества модели переберём в цикле разные варианты глубины дерева и узнаем какая глубина покажет наилучший показатель <b>accuracy</b>.


```python
# обучение модели DecisionTreeClassifier
for depth in range(1, 6):
    model = DecisionTreeClassifier(random_state=12345, max_depth=depth) 
    # обучение модели с заданной глубиной дерева
    model.fit(train_features, train_target)
    # получение предсказаний модели
    valid_predictions = model.predict(valid_features) 
    # расчет качества модели с помощью функции accuracy_score
    result = accuracy_score(valid_target, valid_predictions)

    print("Глубина дерева:", depth,"Accuracy модели:", result) 
```

```python
# поиск наилучших параметров для модели DecisionTreeClassifier
best_model = None
best_result = 0
best_depth = 0
for depth in range(1, 6):
    model = DecisionTreeClassifier(random_state=12345, max_depth=depth)
    # обучение модели с заданной глубиной дерева
    model.fit(train_features, train_target)
    # получение предсказаний модели
    valid_predictions = model.predict(valid_features)
    # расчет качества модели с помощью функции accuracy_score
    result = accuracy_score(valid_target, valid_predictions)
    # определение лучшей оценки качества
    if result > best_result:
        best_model = model
        best_result = result
        best_depth = depth

print("Accuracy наилучшей модели на валидационной выборке:", best_result, "Глубина дерева:", best_depth)
```

Из исследования видим, что на валидацонной выборке наилучшее качество показала модель с глубиной дерева 3, что составляет 0.7854

### Случайный лес

Сохраним в переменную <b>model</b> алгоритм обучения случайный лес. Затем обучим модель. Для оценки качества модели, переберём в цикле разные варианты количества деревьев.

```python
for est in range(10, 51, 10):
    model = RandomForestClassifier(random_state=12345, n_estimators = est) 
    # обучение модели с заданной глубиной дерева
    model.fit(train_features, train_target)
    # получение предсказаний модели
    valid_predictions = model.predict(valid_features) 
    # расчет качества модели с помощью функции accuracy_score
    result = accuracy_score(valid_target, valid_predictions)

    print("Количество деревьев:", est,"Accuracy модели:", result) 
```

Добавим ещё один гиперпараметр глубина дерева и узнаем наилучший показатель <b>accuracy</b>.

```python
# поиск наилучших парамметров для модели RandomForestClassifier
best_model = None
best_result = 0
best_depth = 0
for est in range(10, 61, 10):
    for depth in range(1, 6):
        model = RandomForestClassifier(random_state=12345, n_estimators=est, max_depth=depth)
        # обучение модели с заданной глубиной дерева
        model.fit(train_features, train_target)
        # получение предсказаний модели
        valid_predictions = model.predict(valid_features) 
        # расчет качества модели методом score
        result = model.score(valid_features, valid_target)
        # определение лучшей оценки качества
        if result > best_result:
            best_model = model
            best_result = result
            best_depth = depth
            best_est = est

print("Accuracy наилучшей модели на валидационной выборке:", best_result, "Глубина дерева:", best_depth, 
      "Количество деревьев:", best_est)
```

В моделях случайного леса для исследования мы рассмотрели варианты до 60 деревьев и глубиной 5. Наилучший результат показала модель с глубиной дерева 5 и количеством деревьев 40, качество при этом составило 0.7947 

### Логистическая регрессия

Сохраним в переменную model алгоритм обучения логистическая регрессия. Затем обучим модель. Для оценки качества наилучшей модели зададим алгоритм построения модели 'lbfgs' с количеством итераций 1000.

```python
model = LogisticRegression(random_state=12345, solver='lbfgs', max_iter=1000)
model.fit(train_features, train_target)
print(model.score(valid_features, valid_target))
```

Качество модели логистической регрессии на валидационной выборке показало всего 0.7107.

<b>Вывод</b>

<b>Из проведенного исследования моделей видим, что на валидационной выборке наилучшее качество показала модель случайного леса. Поэтому выберем эту модель для проверки на тестовой выборке.
</b>

## Проверка модели на тестовой выборке

Для проверки модели на тестовой выборке, сохраним в переменную model алгоритм обучения случайный лес. Так как наилучший результат на валидационной выборке показала модель с глубиной дерева 5 и количеством деревьев 40, укажем такие же гиперпараметры на тестовой выборке. Затем обучим модель и узнаем качество на тестовой выборке. 

```python
model = RandomForestClassifier(random_state=12345, n_estimators=40, max_depth=5)
model.fit(train_features, train_target) 

test_predictions = model.predict(test_features)

print(accuracy_score(test_target, test_predictions))
```

<b>Вывод</b>

<b>Оценка качества на тестовой выборке показала 0.78, а это значит, что 8 тарифов из 10 модель предсказывает верно. Однако, 20% ошибок модель всё же допускает, что говорит о необходимости в улучшении качества обучения модели, чтобы достичь наилучших результатов в дальнейшем</b>
